```python
""" 5.  Explain the difference between justify-items and justify-self using code examples.

        Justify-Items:
            justify-items is a property of the grid container that sets the default horizontal alignment for all grid items.

            Example:
            .grid-container {
              display: grid;
              grid-template-columns: repeat(3, 1fr);
              justify-items: center; /* centers all grid items horizontally */
              gap: 10px;
            }

            .grid-item {
              background-color: lightblue;
              padding: 20px;
            }


            <div class="grid-container">
              <div class="grid-item">Item 1</div>
              <div class="grid-item">Item 2</div>
              <div class="grid-item">Item 3</div>
            </div>

            In this example, justify-items: center centers all grid items horizontally within their grid cells.

        Justify-Self:
            justify-self is a property of individual grid items that overrides the default horizontal alignment set by justify-items.

            Example:
            .grid-container {
              display: grid;
              grid-template-columns: repeat(3, 1fr);
              justify-items: center; /* centers all grid items horizontally */
              gap: 10px;
            }

            .grid-item {
              background-color: lightblue;
              padding: 20px;
            }

            .item1 {
              justify-self: start; /* overrides default alignment for Item 1 */
            }

            .item3 {
              justify-self: end; /* overrides default alignment for Item 3 */
            }


            <div class="grid-container">
              <div class="grid-item item1">Item 1</div>
              <div class="grid-item">Item 2</div>
              <div class="grid-item item3">Item 3</div>
            </div>

            In this example:
            - justify-items: center centers all grid items horizontally by default.
            - justify-self: start overrides the default alignment for Item 1, aligning it to the start of its grid cell.
            - justify-self: end overrides the default alignment for Item 3, aligning it to the end of its grid cell. """
```
